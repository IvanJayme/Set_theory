# Providing Identities
   
**a. $A \setminus B = A \cap B^c$**
   
**Given:** Sets \(A\) and \(B\).

**To Prove:** $A \setminus B = A \cap B^c \$.

**Proof:**
For any element \( x \):
1. Assume $x \in A \setminus B$:

   This means $x$ is in $A$ but not in $B$.

   Thus, $x$ is in $A \cap B^c$.

2. Assume $x \in A \cap B^c$:

   This means $x$ is in both $A$ and the complement of $B$.

   Thus, $x$ is in $A \setminus B$.

Since $x$ belonging to one side implies it belonging to the other, we conclude $A \setminus B\cap B^c$.=\n
 
**Venn diagram**

```python
A = {'a', 'b', 'c', 'd'}
B = {'c', 'd', 'e', 'f'}
venn = venn2([A, B], ('A', 'B'))
venn.get_label_by_id('10').set_text('A - B')
venn.get_label_by_id('01').set_text('B - A')
venn.get_label_by_id('11').set_text('A $\cap$ B')

plt.title('Venn Diagram: A \\ B = A $\cap B^c$')
plt.show()
```


**b. B = (B $\cap$ A) $\cup$ (B $\cap$ $A^{c}$)**

**Given:** Set $B$.

**To Prove:** $B = (B \cap A) \cup (B \cap A^c)$.

**Proof:**

For any element $x$:

1. Assume $x \in B$:

   This means $x$ is in $B$.

   Thus, $x$ is in $(B \cap A) \cup (B \cap A^c)$.

2. Assume $x \in (B \cap A) \cup (B \cap A^c)$:

   This means $x$ is in either $B \cap A$ or $B \cap A^c$.

   If $x$ is in $B \cap A$, it's in $B$.

   If $x$ is in $B \cap A^c$, it's in $B$.

   Therefore, $x$ is in $B$.

Since $x$ belonging to one side implies it belonging to the other, we conclude $B = (B \cap A) \cup (B \cap A^c)$.

**Venn diagram**

```python
A = {'a', 'b', 'c', 'd'}
B = {'c', 'd', 'e', 'f'}

venn = venn2([A, B], ('A', 'B'))
venn.get_label_by_id('10').set_text('B $\cap A^c$')
venn.get_label_by_id('01').set_text('B $\cap$ A')
venn.get_label_by_id('11').set_text('A $\cap$ B')

plt.title('Venn Diagram: B = (B $\cap$ A) $\cup$ (B $\cap$ $A^c$)')
plt.show()
```

**c. B\\A = B $\cap$ $A^{c}$**

**Given:** Set $B$.

**To Prove:** $B \setminus A = B \cap A^c$.

**Proof:**

For any element $x$:

1. Assume $x \in B \setminus A$:

   This means $x$ is in $B$ but not in $A$.

   Thus, $x$ is in $B \cap A^c$.

2. Assume $x \in B \cap A^c$:

   This means $x$ is in $B$ and in the complement of $A$.

   Thus, $x$ is in $B \setminus A$.

Since $x$ belonging to one side implies it belonging to the other, we conclude $B \setminus A = B \cap A^c$.

```python
A = {'a', 'b', 'c', 'd'}
B = {'c', 'd', 'e', 'f'}
U = set.union(A, B)  # Universal set

difference_B_A = B - A
intersection_B_complement_A = B.intersection(U - A)
venn2([A, B], ('A', 'B'))

plt.text(0.25, 0.5, len(intersection_B_complement_A), color='white')
plt.text(0.7, 0.5, len(A - B), color='white')
plt.text(0.45, 0.2, len(difference_B_A), color='white')

plt.title('Venn Diagram: B \ A = B $\cap$ $A^c$')
plt.show()

print("Is B \\ A equal to B $\cap$ A^c", difference_B_A == intersection_B_complement_A)
```

**d. A $\cup$ B = A $\cup$ (B $\cap$ $A^{c}$)**

**Given:** Sets $A$ and $B$.

**To Prove:** $A \cup B = A \cup (B \cap A^c)$.

**Proof:**

For any element $x$:

1. Assume $x \in A \cup B$:

   This means $x$ is in either $A$ or $B$.

   If $x$ is in $A$, it's also in $A \cup (B \cap A^c)$.

   If $x$ is in $B$, two cases arise:
   
   a. If $x$ is also in $A$, it's in $A \cup (B \cap A^c)$.
   
   b. If $x$ is not in $A$, it's in $B \cap A^c$, and thus in $A \cup (B \cap A^c)$.

2. Assume $x \in A \cup (B \cap A^c)$:

   This means $x$ is in either $A$ or in the intersection of $B$ and the complement of $A$.

   If $x$ is in $A$, it's also in $A \cup B$.

   If $x$ is in $B \cap A^c$, it's in $B$ and not in $A$ (by definition). Thus, it's in $A \cup B$.

Since $x$ belonging to one side implies it belonging to the other, we conclude $A \cup B = A \cup (B \cap A^c)$.

```python
A = {'a', 'b', 'c', 'd'}
B = {'c', 'd', 'e', 'f'}
venn = venn2([A, B], ('A', 'B'))
venn.get_label_by_id('10').set_text('B $\cap A^c$')
venn.get_label_by_id('01').set_text('B $\cap$ A')
venn.get_label_by_id('11').set_text('A $\cap$ B')

plt.title('Venn Diagram: A $\cup$ B = A $\cup$ (B $\cap$ $A^c$)')
plt.show()
```

**e. $(\textbf{U}_{i=1}^{\\infty} A_{i}^{c})^{c}$ = $\cap_{i=1}^{\\infty}A_{i}$**


**Given:** Sets $A_i$ for $i = 1, 2, \ldots$.

**To Prove:** $(\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c} = \cap_{i=1}^{\infty} A_{i}$.

**Proof:**

An element $x$ belongs to $(\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c}$ if and only if it does not belong to the union of all complements of $A_i$.

1. Assume $x \in (\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c}$:

   This means $x$ is not in the union of all complements of $A_i$. In other words, $x$ is not in any of $A_{i}^{c}$.

   Thus, $x$ belongs to each $A_i$.

2. Assume $x \in \cap_{i=1}^{\infty} A_{i}$:

   This means $x$ belongs to each $A_i$.

   Therefore, $x$ is not in the union of all complements of $A_i$.

Since $x$ belonging to one side implies it belonging to the other, we conclude $(\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c} = \cap_{i=1}^{\infty} A_{i}$.

**f. |A $\cup$ B| = |A| + |B| - |A $\cap$ B|**

**Given:** Sets $A$ and $B$.

**To Prove:** $|A \cup B| = |A| + |B| - |A \cap B|$.

**Proof:**

Starting with the left side, $|A \cup B|$, this represents the size of the set obtained by combining elements from both $A$ and $B$. However, this count may include elements that appear in both sets, leading to overcounting.

On the right side, $|A| + |B| - |A \cap B|$, the first two terms represent the sum of the sizes of sets $A$ and $B$. However, this sum overcounts elements that are present in the intersection $A \cap B$, as they are counted in both $A$ and $B$.

The term $|A \cap B|$ represents the size of the set of elements that are in both $A$ and $B$. To avoid double-counting, we subtract this term once from the sum of individual sizes.

Thus, by subtracting $|A \cap B|$ from the sum, we ensure that each element is counted correctly in the union $A \cup B$.

Hence, we conclude that $|A \cup B| = |A| + |B| - |A \cap B|$.

# Finite Sets and The Counting Principles

In a Survey of 120 People, it was found that:


- 65 read Newsweek Magazine
- 45 read Times Magazine
- 42 read Fortune Magazine
- 20 read Both Newsweek and Time
- 25 read both Newsweek and Fortune
- 15 read both Time and Fortune
- 8 read all three magazines

With these information do the following:

1. Draw the venn diagram using python.
2. Find the number of people who read at **least one** of the three magazines.
3. Find the number of people who read exactly one magazine.

```python
# Given data
total_people = 120
newsweek_readers = 65
times_readers = 45
fortune_readers = 42
both_newsweek_times = 20
both_newsweek_fortune = 25
both_times_fortune = 15
all_three = 8

# Calculate the number of people who read exactly one magazine
only_newsweek = newsweek_readers - both_newsweek_times - both_newsweek_fortune + all_three
only_times = times_readers - both_newsweek_times - both_times_fortune + all_three
only_fortune = fortune_readers - both_newsweek_fortune - both_times_fortune + all_three

# Draw the Venn diagram
venn = venn3(subsets=(only_newsweek, only_times, both_newsweek_times, only_fortune, both_newsweek_fortune, both_times_fortune, all_three),
             set_labels=('Newsweek', 'Time', 'Fortune'))

# Display the diagram
plt.title("Venn Diagram of Magazine Readers")
plt.show()

# Calculate the number of people who read at least one of the three magazines
read_at_least_one = newsweek_readers + times_readers + fortune_readers - (both_newsweek_times + both_newsweek_fortune + both_times_fortune) + all_three
print("Number of people who read at least one magazine:", read_at_least_one)

# Calculate the number of people who read exactly one magazine
total_exactly_one = only_newsweek + only_times + only_fortune
print("Number of people who read exactly one magazine:", total_exactly_one)
```
